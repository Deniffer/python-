````python

import simpy 
import random
RANDOM_SEED=22
num_road = 2 #道路数量
num_park = 4 #泊车位数量
cost_time=random.uniform(15,45) #上下车所花费时间服从均匀分布
simulate_time=3600 #总仿真时间
behind_car=random.uniform(15,25)
class Parking_Lot(object):
    ##设定泊车位数量，上客时间
    def __init__(self,env,num_parks,time):
        self.env = env
        self.park = simpy.Resource(env,num_parks)
        self.time =time
    def get_on(self,car):
        yield self.env.timeout(random.uniform(15,45))
        print('上车花费了%.2f'%random.uniform(15,45))
def car(env,name,pl):
    print('%s 正在上客，距离第一辆车上客已经过去%.2fs.'%(name,env.now))
    with pl.park.request() as request:#上客
        yield env.timeout(random.uniform(15,45))
        print('%s 进入泊车位并上客%.2f'%(name,env.now))
        yield env.process(pl.get_on(name))
        print('%s 离开泊车位，时间%.2fs'%(name,env.now))
def set_up(env,num_parks,time,BEHIND_car):
    parking_lot = Parking_Lot(env,num_parks,time)
    for i in range(num_park):
        env.process(car(env,'Car%d'%i,parking_lot))
    while True:
        yield env.timeout(random.uniform (20,30))
        i += 1
        env.process(car(env,'Car%d'%i,parking_lot))
#
print('各车到达泊车位，开始上客')
random.seed(RANDOM_SEED) #帮助刷新测试结果
env = simpy.Environment()
env.process(set_up(env,num_park,cost_time,behind_car))
#
env.run(until=simulate_time)
````

## 图形绘制的一般概念<br>
<font face='kaiti' size=5 style='line-height:1.5'>
    
<img src="anatomy.png" width=640 height=640 alt="图形对象解析" align=center>
&emsp;&emsp;Matplotlib绘制的图形都包含了几个对象。分别是
- Figure对象：整个图形，记录了所有Axes，一些 “特殊” 的Artist（标题，图形图例等）和画布canvas。可以用gcf()获取当前图形，clf()清除当前图形。
- Axes对象：具有数据空间的图像区域，一个图形窗中可以有多个Axes对象，可以用subplot()或axes()建立，也可以用gca()获取当前Axes对象，cla()清除当前Axes对象。每个Axes对象都有一个标题，一个x标签和一个y标签。Axes还包含两个（或3D的三个）Axis对象。Axes对象有大量的方法，可以用于绘制曲线，设置标签，标题，网格线，等等，可以用于对图形中的对象做底层控制。
- Axis对象：坐标轴的数字特征，负责设置图形范围并生成刻度线和标记符。
- Artist对象：基本上图上看到的一切都是Artist对象（甚至是图，轴和轴对象）。这包括Text对象，Line2D对象，集合对象，Patch对象等。渲染图形时，所有艺术家都被绘制到画布（canvas）上。

&emsp;绘制图形和设置对象的属性可以使用这些对象的相应方法。但一些简单的功能都已集成在pyplot中。

## 简单绘图与设置
<font face='kaiti' size=5 style='line-height:1.5'>
&emsp;&emsp;使用pyplot子模块可以完成大部分的绘图任务，并且可以对很多图形属性进行调整，如：图片大小和分辨率（dpi）、线宽、颜色、风格、坐标轴、坐标轴以及网格的属性、文字与字体属性等。pyplot模块中常用的绘图函数如下：
    
|函 数|说 明|
|:--|:--|
|plt.plot()|绘制线性二维图，折线图，散点图|
|plt.scatter()|绘制散点图|
|plt.bar|绘制条形图|
|plt.pie()|绘制饼型图|
|plt.hist()|绘制二维直方图，可显示数据的分配情形|
|plt.boxplot()|绘制样本数据的箱型图|

&emsp;&emsp;注：以上绘图函数要求输入的x,y数据是np.arrays类型。下面先以线图为例，结合说明绘图中的各种控制选项。

### 线图 plt.plot()
<font face='kaiti' size=5 style='line-height:1.5'>
&emsp;&emsp;使用格式：plt.plot(x, y, S)，plt.plot(x1, y1, S1, x2, y2, S2,...). 
用于绘制x为横坐标，y为纵坐标的函数图形，字符串参数S指定绘制时的线型、点的样式、颜色。

|线型字符|字符含义|点的样式|字符含义|
|:---|:---|:---|:---|
|-|实线|.|圆点|
|--|虚线|o|圆圈|
|-.|点划线|`*`|星号|
|:|点|+|加号|
|**颜色字符**|**字符含义**|<|向左的三角形|
|b|蓝色blue|>|向右的三角形|
|g|绿色green|^|向上的三角形|
|r|红色red|v|向下的三角形|
|c|青色cyan|s|正方形|
|m|品红magenta|d|菱形|
|y|黄色yellow|p|五角星|
|k|黑色black|h|六边形|
|w|白色white|x|x形状（叉）|

````python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline

X = np.linspace(-np.pi, np.pi, 51,endpoint=True)
C,S = np.cos(X), np.sin(X)

# 创建一个 7 * 4 inches 的图，并设置分辨率为 80
plt.figure(figsize=(7,4), dpi=80)

# 绘制余弦曲线，使用蓝色的实线，圆圈
plt.plot(X, C, 'bo-')   # 'bo-'分别对应'[color][marker][line]'

# 绘制正弦曲线，使用红色点线，菱形
plt.plot(X, S, 'rd:')

plt.show()
````

### pyplot设置坐标轴范围plt.axis()，添加网格 plt.grid()<br>
<font face='kaiti' size=5 style='line-height:1.5'>
 
- plt.axis()函数用于获取或设置坐标轴特性，与MATLAB的用法相似。
    - plt.axis()返回当前坐标轴的范围 `[xmin， xmax, ymin, ymax]`
    - plt.axis([xmin， xmax, ymin, ymax])用于设置当前坐标轴范围
    - plt.axis('off') 不显示坐标轴及坐标轴标签
    - plt.axis('equal') 设置不同坐标轴的单位长度相同
    - plt.axis('tight') 改变坐标轴的范围，使数据能全部显示，并居中
    - plt.axis('square') 设置坐标轴的取值宽度相等，单位长度相同，绘图区域为正方形
- plt.grid(b=None, which='major', axis='both', **kwargs)
    - b 取布尔值，用于绘图区域显示或不显示网格
    - b为空，且len(kwargs)==0，则plt.grid()用于切换网格的显示状态
    - which可取'major'或'minor'，显示稀疏或细密的网格
    - axis可取'both','x'或'y'，控制显示哪些网格线
    - kwargs以关键字参数的方式控制网格线的特性，如ax.grid(color='r', linestyle='-', linewidth=2)

````python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline

X = np.linspace(-np.pi, np.pi, 201,endpoint=True)
C,S = np.cos(X), np.sin(X)

plt.figure(figsize=(7,4), dpi=80)
plt.axis([-3.5, 3.5, -1.1, 1.1]) # 设置坐标轴范围
plt.axis('equal') # 使坐标轴单位长度相同，可能改变已设置的坐标轴范围
plt.grid('on',color='m', linestyle=':')  # 设置网格线
plt.plot(X, C, color="cyan", linewidth=2.5, linestyle="-")
plt.plot(X, S, color="orange", linewidth=2.5, linestyle="-")

plt.show()
````

## 多子图<br>
<font face='kaiti' size=5 style='line-height:1.5'>
    
&emsp;&emsp;pyplot在一个图形窗口中可以设置多个坐标系，也就是可以建立多个绘图子区域。可以通过plt.subplot()函数设置子图的区域，而使用plt.axes()可以更灵活地设定子图的位置。

### plt.subplot()

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x1 = np.linspace(0.0, 5.0, 101, endpoint=True)
x2 = np.linspace(0.0, 2.0, 101, endpoint=True)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.figure(figsize=(8,6), dpi=80)
plt.subplot(2, 1, 1)  # 上下2个子图，子图1
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)  # 子图2
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
````

### plt.axes()
<font face='kaiti' size=5 style='line-height:1.5'>
    
&emsp;&emsp;plt.axes((left, bottom, width, height))用于添加一个坐标系到当前图像窗口中，(left, bottom, width, height)为图形窗口中的相对位置，取值在[0,1]上。<br>
&emsp;&emsp;plt.axes()还可以放在plt.subplot()后面一起使用。

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x1 = np.linspace(0.0, 5.0, 101, endpoint=True)
x2 = np.linspace(0.0, 2.0, 101, endpoint=True)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(223)  # 子图1
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.axes([0.45,0.45,0.55,0.55])  # 子图2
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
````

Matplotlib中产生动态图的最方便地途径是使用Animation类。

&emsp;&emsp;下面的例子演示雨滴落在平面上的动态效果图，来自https://github.com/rougier/matplotlib-tutorial/blob/master/scripts/rain-static.py <br>
    如果notebook中不能正常显示动画的话，可以copy到编辑器中保存再运行。

````python
# -----------------------------------------------------------------------------
# Copyright (c) 2015, Nicolas P. Rougier. All Rights Reserved.
# Distributed under the (new) BSD License. See LICENSE.txt for more info.
# -----------------------------------------------------------------------------
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# No toolbar
mpl.rcParams['toolbar'] = 'None'

# New figure with white background
fig = plt.figure(figsize=(6,6), facecolor='white')

# New axis over the whole figureand a 1:1 aspect ratio
# ax = fig.add_axes([0,0,1,1], frameon=False, aspect=1)
ax = fig.add_axes([0.005,0.005,0.990,0.990], frameon=True, aspect=1)

# Number of ring
n = 50
size_min = 50
size_max = 50*50
          
# Ring position 
P = np.random.uniform(0,1,(n,2))

# Ring colors
C = np.ones((n,4)) * (0,0,0,1)

# Alpha color channel goes from 0 (transparent) to 1 (opaque)
C[:,3] = np.linspace(0,1,n)

# Ring sizes
S = np.linspace(size_min, size_max, n)

# Scatter plot
scat = ax.scatter(P[:,0], P[:,1], s=S, lw = 0.5,
                  edgecolors = C, facecolors='None')

# Ensure limits are [0,1] and remove ticks
ax.set_xlim(0,1), ax.set_xticks([])
ax.set_ylim(0,1), ax.set_yticks([])


def update(frame):
    global P, C, S

    # Every ring is made more transparent
    C[:,3] = np.maximum(0, C[:,3] - 1.0/n)

    # Each ring is made larger
    S += (size_max - size_min) / n

    # Reset ring specific ring (relative to frame number)
    i = frame % 50
    P[i] = np.random.uniform(0,1,2)
    S[i] = size_min
    C[i,3] = 1

    # Update scatter object
    scat.set_edgecolors(C)
    scat.set_sizes(S)
    scat.set_offsets(P)
    return scat

animation = FuncAnimation(fig, update, interval=10)
# animation.save('../figures/rain.gif', writer='imagemagick', fps=30, dpi=72)
plt.show()
````
