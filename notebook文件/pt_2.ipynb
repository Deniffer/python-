```python
import re
def get_clean_words(text):
    result=[]
    result_fin=[]
    with open(text,encoding="utf-8")as key:
        for i in key:
            result.append(i)
    for i in range(len(result)):
        result_all=[]
        result_clean=re.findall("\".*?\"",result[i])
        for j in range(len(result_clean)):
            result_all.append(eval(result_clean[j]))
        result_fin.append(result_all)
    return result_fin
def get_clean_weight(text):
    num=[]
    all_num=[]
    with open(text,encoding="utf-8")as key_2:
        for i in key_2:
            num.append(i)
    for i in range(len(num)):
        all_num.append(re.findall("\..*?\*",num[i]))
    all_clean_num=[]
    for i in range(30):
        clean_num=[]
        for j in all_num[i]:
            j=j.strip(",")
            j=float(j.strip("*"))
            clean_num.append(j)
        clean_num=[i*(1/sum(clean_num)) for i in clean_num]
        all_clean_num.append(clean_num)
    len(all_clean_num[0])
    return all_clean_num
```
