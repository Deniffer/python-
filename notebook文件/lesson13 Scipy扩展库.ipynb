## 插值 interpolate<br>
<font face='kaiti' size=5 style='line-height:1.5'>
    
&emsp;&emsp;interpolate模块提供了多个用于插值的函数，可以用于单变量插值，多变量插值，B样条插值等等。以下简单介绍几个常用的插值函数的用法和实例。

更多函数，见 https://docs.scipy.org/doc/scipy/reference/interpolate.html  <br>
详细用法，见 https://docs.scipy.org/doc/scipy/reference/tutorial/interpolate.html

### 一维插值 interp1d()<br>
<font face='kaiti' size=5 style='line-height:1.5'>

&emsp;interp1d(x, y[, kind='linear', **args])
- x, y: 插值节点的x, y取值；
- kind: 插值方法，有'linear', 'nearest', 'previous', 'next', 'zero', 'slinear', 'quadratic', 'cubic'，其中'linear', 'nearest', 'previous', 'next'分别是线性插值，最近邻插值，向前插值，向后插值。'zero', 'slinear', 'quadratic'以及'cubic'分别是零次，一次，二次，三次样条插值。
- 返回： 一个插值函数。

````python

import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
% matplotlib inline
x = np.linspace(0, 10, num=11, endpoint=True)
y = np.cos(-x**2/9.0)
f = interp1d(x, y)   # 默认线性插值
f2 = interp1d(x, y, kind='cubic')   # 三次样条插值

xnew = np.linspace(0, 10, num=41, endpoint=True)

plt.plot(x, y, 'o', xnew, f(xnew), '-', xnew, f2(xnew), '--')
plt.legend(['data', 'linear', 'cubic'], loc='best')
plt.show()
````

### 多维规则网格插值 interpn()<br>
<font face='kaiti' size=5 style='line-height:1.5'>

interpn(points, values, xi, method='linear', bounds_error=True, fill_value=nan)
- points: 浮点数numpy数组构成的元组，用于定义多维规则网格
- values: 数组类型，多维规则网格点的取值
- xi: 多维数组，表示插值网格内的采样点坐标
- method： 插值方法，可选'linear', 'nearest', 'splinef2d'。这里'splinef2d'只支持二维数据
- bounds_error: 采样点越界处理，'True'表示抛出异常信息，'False'则使用fill_value填充。
- fill_value: 采样点越界时的填充取值，如果为None，则越界点做外插处理。
- 返回值： 采样点所得插值。

````python
% matplotlib inline
import numpy as np
from scipy.interpolate import interpn
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 插值节点
x = np.linspace(-4, 4, 17)
X, Y = np.meshgrid(x, x)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
# 做二维插值
xi = np.linspace(-4, 4, 81)
(Xi, Yi) = np.meshgrid(xi, xi)
Zi = interpn((x, x), Z, (Xi, Yi), method='splinef2d')

# 绘图
fig = plt.figure(figsize=(10, 5))
ax = Axes3D(fig)
ax.plot(X.flatten(), Y.flatten(), Z.flatten(), 'r.')
ax.plot_surface(Xi, Yi, Zi, rstride=1, cstride=1, cmap=plt.cm.winter)
plt.show()
````

### 多维非网格数据插值<br>
<font face='kaiti' size=5 style='line-height:1.5'>

&emsp;griddata(points, values, xi, method='linear', fill_value=nan, rescale=False)


````python
% matplotlib inline
import numpy as np
from scipy.interpolate import griddata
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 插值节点
np.random.seed(19491001)   # 指定随机数种子
points = np.random.rand(200,2)*8-4
X = points[:,0]
Y = points[:,1]
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
# 做二维插值
xi = np.linspace(-4, 4, 81)
(Xi, Yi) = np.meshgrid(xi, xi)
Zi = griddata((X, Y), Z, (Xi, Yi), method='cubic')

# 绘图
fig = plt.figure(figsize=(10, 5))
ax = Axes3D(fig)
ax.plot(X, Y, Z, 'r.')
ax.plot_surface(Xi, Yi, Zi)
plt.show()
````

### 一维样条插值<br>
<font face='kaiti' size=5 style='line-height:1.5'>

- splrep(x, y[, w=None, xb=None, xe=None, k=3, s=None, ...]) 由给定的数据点x, y求k次B样条逼近函数的表达式。w为权值数组，xb，xe分别为样条拟合的区间，s度量光滑性，越大越光滑
- splev(x, tck, der=0, ext=0): 求B样条函数值或它的导数值。x为待求函数值的点列数组，tck为splrep的返回，der为导数的阶数，ext为越界点的处理策略。

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
%matplotlib inline

x = np.linspace(0, 2*np.pi, 9)
y = np.sin(x)
tck = interpolate.splrep(x, y, s=0)  # 
xnew = np.linspace(0, 2*np.pi, 101)
ynew = interpolate.splev(xnew, tck, der=0)  # 

plt.figure(figsize=(15, 5))
plt.rcParams['font.sans-serif'] = 'SimHei'  # 设置中文字体
plt.rcParams['axes.unicode_minus'] = False  # 设置正常显示负号
plt.subplot(121)
plt.plot(x, y, 'b-x', xnew, ynew, xnew, np.sin(xnew),'--')
plt.legend(['线性插值', '三次样条', '真实值'])
plt.axis([-0.05, 6.33, -1.05, 1.05])
plt.title(u'3次样条插值')

# Derivative of spline

yder = interpolate.splev(xnew, tck, der=1)
plt.subplot(122)
plt.plot(xnew, yder, xnew, np.cos(xnew),'--')
plt.legend(['三次样条', '真实值'])
plt.axis([-0.05, 6.33, -1.05, 1.05])
plt.title(u'样条插值的导数估计')
plt.show()

````

### 由函数表达式求积分<br>
<font face='kaiti' size=5 style='line-height:1.5'>

- quad(func, a, b[, args=(), full_output=0, ...]) : 求[a,b]上函数func的定积分，返回积分值和误差估计。args为func中的其他参数，full_output若非零，则返回积分信息的字典。
- dblquad(func, a, b, gfun, hfun[, args=(), ...]) : 求函数func(y, x)在x in [a, b] 以及 y in [gfun(x), hfun(x)]的二重积分。
- tplquad(func, a, b, gfun, hfun, qfun, rfun[, args=(), ...]): 三重积分。

<font face='kaiti' size=5 style='line-height:1.5'>
&emsp;求定积分 $\int_1^\infty\frac{e^{-xt}}{t^n} dt$ 在 n =3, x=1.0的值。

````python
from scipy.integrate import quad

def integrand(t, n, x):
    return np.exp(-x*t) / t**n

s = quad(integrand, 1, np.inf, args=(3, 1.0))[0]
s
````

### 由采样点列求数值积分<br>
<font face='kaiti' size=5 style='line-height:1.5'>
    
- trapz(y, x=None, dx=1.0, axis=-1): 梯形方法，由y(x)点列沿axis轴求积分，dx表示x为None时的采样间隔。
- cumtrapz(y, x=None, dx=1.0, axis=-1, initial=None): 梯形方法求累计积分。
- simps(y, x=None, dx=1, axis=-1, even='avg')：辛普森方法，由y(x)点列沿axis轴求积分，even为积分区间两头的处理方法。

<font face='kaiti' size=5 style='line-height:1.5'>
&emsp;辛普森方法求积分 $\int_1^4 x^2 dx$ 和 $\int_1^4 x^3 dx$

````python
import numpy as np

def f1(x):
    return x**2

def f2(x):
    return x**3

x = np.linspace(1,4,31)

from scipy.integrate import simps
I1 = simps(f1(x), x)
print(I1)

I2 = simps(f2(x), x)
print(I2)
````

<font face='kaiti' size=5 style='line-height:1.5'>
&emsp;绘制标准正态分布的分布函数$\frac1{\sqrt{2\pi}}\int_{-\infty}^{+\infty}e^{-\frac{x^2}2}dx $的曲线

````python
import numpy as np
from scipy.integrate import cumtrapz
from matplotlib import pyplot as plt
% matplotlib inline

def f(x):
    return np.exp(-x*x/2)/np.sqrt(2*np.pi)

x = np.linspace(-5, 5, 101)
y = f(x)  # 概率密度函数值
f = cumtrapz(y, x, initial=0)  # 分布函数值

fig = plt.figure(figsize=(10,5))
plt.plot(x,y,x,f)
plt.legend(('pdf', 'cdf'))
plt.plot(0, 0.5,'ro', [0,0],[0,1])
plt.show()
````

### 微分方程初值问题 <br>
<font face='kaiti' size=5 style='line-height:1.5'>

&emsp;&emsp;一阶微分方程可表示为 $ y' = f (y, t)$。二阶及更高阶微分方程可转化为一阶微分方程组求解。常用的函数如下：<p>
&emsp;solve_ivp(fun, t_span, y0[, method='RK45', t_eval=None, ...]): 
- fun用于定义方程$ y ' = f (y, t)$的右边部分
- t_span是一个求解区间(t0, tf)，y0是t0处的函数值
- method是求解方法，可选'RK45', 'RK23', 'Radau', 'PDF', 'LSODA'.
- t_eval是t的取值向量，必须单调且在区间[t0, tf]上。

<font face='kaiti' size=5 style='line-height:1.5'>
【例】方程 $ y' + \frac y x = \frac{\sin x}x,$&emsp;$ y|_{x=\frac{\pi}2}=0$ 的解是 $ y = -\frac{\cos x} x$。试用scipy求该方程的数值解，并绘图与精确解比较。

````python
import numpy as np
from scipy.integrate import solve_ivp
from matplotlib import pyplot as plt
% matplotlib inline

def fun(x,y):
    return -y/x + np.sin(x)/x

x_span = (np.pi/2, np.pi*3)
y0 = [0]
x_eval = np.linspace(np.pi/2, np.pi*3, 101)
solve = solve_ivp(fun, x_span, y0, t_eval=x_eval )
solve.y.shape = solve.t.shape
y_true = -np.cos(solve.t)/solve.t

fig = plt.figure(figsize=(8,5))
plt.plot(solve.t, solve.y,'r-', solve.t, y_true,'b--')
plt.legend(('数值解','真实解'))
plt.show()
````

|moment| non-central moments of the distribution |该分布的非中心距|

&emsp;<font face='kaiti' size=5>以下以正态分布为例，说明其中一些方法的应用。

<font face='kaiti' size=5 style='line-height:1.5'> 
【例】已知一个随机变量$X\sim N(2.5, 1.2^2)$，绘制它的概率密度函数和累积分布函数的图形，求$P(1.5\leq X\leq 3)$，并在概率密度函数的图形上用面积表示该概率。又已知 $ P(X<a) = 0.8$, 求$a$的值，并在概率密度函数曲线上标注该点。

````python
import numpy as np
from scipy.stats import norm
from matplotlib import pyplot as plt

N = norm(loc=2.5, scale=1.2)   # 产生该正态分布随机变量

# 绘制概率密度函数和累积分布函数的图形
fig = plt.figure(figsize=(10,5))
x = np.linspace(-1, 6, 201)
pdf = N.pdf(x)  # 概率密度函数
cdf = N.cdf(x)  # 累积分布函数
plt.plot(x, pdf, x, cdf, x, x*0)
plt.legend(['pdf','cdf'])
# 求 P(1.5≤X≤3)，并在概率密度函数的图形上用面积表示该概率
P1 = N.cdf(1.5)  # 用分布函数求概率
P2 = N.cdf(3)
print('P(1.5≤X≤3) =', P2-P1)
# 以下绘制填充图
x1 = np.linspace(1.5, 3, 31)
pdf1 = N.pdf(x1)
plt.fill_between(x1, 0*x1, pdf1, facecolor='green', alpha = 0.3)
plt.text(1.6, 0.5*pdf1[0], 'P(1.5≤X≤3) = \n'+ str(P2-P1))
# 已知 P(X<a)=0.8, 求 a 的值，并在概率密度函数曲线上标注该点
a = N.ppf(0.8)  # 求分位数
print('P(X<a)=0.8, a =', a)
pa = N.pdf(a)
plt.plot(a, pa, 'ro')
plt.text(a+0.1, pa, 'P(X<'+ str(a)+ ') = 0.8', color = 'r')
# 绘制图形
plt.show()
````

【例】对一组服从正态分布的随机样本进行描述性统计，并做参数估计。

````python
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline

# 产生随机数，做描述性统计
x = stats.norm.rvs(2.5, 1.2, size=10000)
st = stats.describe(x)
print(st)
# 绘制直方图和概率密度函数
fig = plt.figure(figsize=(10,5))
plt.hist(x, 50, density=True, facecolor='g', alpha=0.75) # 直方图

x1 = np.linspace(x.min(), x.max(), 201)
pdf = stats.norm.pdf(x1, loc=2.5, scale=1.2)  # 理论上的概率密度函数
plt.plot(x1, pdf,'r') 
# 参数估计
mean, std = stats.norm.fit(x)  # 返回均值，标准差
pdf1 = stats.norm.pdf(x1, loc=mean, scale=std)  # 估计的概率密度函数
plt.plot(x1, pdf1, 'b--')
plt.legend(('真实的概率密度函数','估计的概率密度函数'))
plt.grid(True)
plt.show()
````

<font face='kaiti' size=5 style='line-height:2'>
【例】蒙特卡洛方法求积分 $ I = \int_{-\infty}^{\infty}xe^{-\frac12 x^2}dx $

<font face='kaiti' size=5 style='line-height:2'>
    
&emsp;分析：取辅助函数 $ p(x) = \frac1{\sqrt{2\pi}}e^{-\frac12 x^2}$，这是标准正态分布的概率密度函数。对$p(x)$求数学期望，得：<br>
<center> $\mu = \int_{-\infty}^{\infty}x p(x)dx = \int_{-\infty}^{\infty}x\cdot \frac1{\sqrt{2\pi}} e^{-\frac12 x^2}dx$ </center> 
则$ I = \sqrt{2\pi}\mu$，取$n$个标准正态随机数$u_i$，有 $ I\approx \sqrt{2\pi} \frac1n \sum u_i $

````python
import numpy as np
from scipy.stats import norm

x = norm.rvs(0, 1.0, size=10000)
I = np.sqrt(2*np.pi)*np.mean(x)
print(I)
````

````python
# 梯形法求积分，以[-100,100]上的积分值代替(-inf, inf)上的积分
import numpy as np

x = np.linspace(-100,100, 10001)
dx = 200/10000  
f = x * np.exp(-x*x/2)
I = np.sum(f[0:-1] + f[1:])/2*dx
print(I)
````
